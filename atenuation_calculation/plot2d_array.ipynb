```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.random.randint(256, size=(10, 10)) #define la matriz o el vector para graficar

fig = plt.figure(figsize=(8,6)) #define el canva
plt.imshow(X, cmap="inferno")                   #grafica 

#El método matplotlib.pyplot.imshow() toma un array 2D como entrada y renderiza el array 
#dada como una imagen rasterizada.
#Por defecto, los valores son mapeados usando el mapa de colores viridis.

plt.title("Plot 2D array")      #coloca el título
plt.show()                      #vuelve a grafica
```

```python
X=np.random.randint(300, size=(10, 10)) # 300 valor máximo, 10 número de filas, 30 número de columnas

fig = plt.figure(figsize=(10,10)) #columnas y filas del array
plt.pcolormesh(X,cmap="plasma")
plt.title("Plot 2D array")
plt.colorbar()
plt.show()
```

```python
array1 = np.array([
    
    
    [2,3],

                           [8, 9],

                           [10,
                            19]

])
print(array1)

array2 = np.zeros((10,10))
array3 = np.identity(10)
print(type(array3))

fig = plt.figure(figsize=(10,10)) #columnas y filas del array
plt.pcolormesh(array1,cmap="plasma")
plt.title("Plot 2D array")
plt.colorbar()
plt.show()

```

```python
a = np.array((1,22,3))

b = np.array((2,3,4))
print(a)
c =np.column_stack((a,b))

print(type(c))

fig = plt.figure(figsize=(3,2)) #columnas y filas del array
plt.pcolormesh(c,cmap="plasma")
plt.title("Plot 2D array")
plt.colorbar()
plt.show()

```
